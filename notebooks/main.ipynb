## Librerias 

```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from pathlib import Path #Solucion para encontrar la ruta del archivo sin importar el sistema operativo
```

## Carpeta de imagenes

```python
IMAGES_PATH = "Galeria"  # Nombre de la carpeta
os.makedirs(IMAGES_PATH, exist_ok=True) # Crea la carpeta si no existe

def save_fig(fig_id, tight_layout=True, fig_extension="pdf", resolution=300):
    path = os.path.join(IMAGES_PATH, fig_id + "." + fig_extension)
    print("Imagen guardada como ", fig_id, "en la carpeta", IMAGES_PATH) #Mensaje mas informativo
    if tight_layout:
        plt.tight_layout()
    plt.savefig(path, format=fig_extension, dpi=resolution)
```

## Carga de los datos

```python
BASE_DIR = Path().resolve()

while not (BASE_DIR / "datos").exists():
    BASE_DIR = BASE_DIR.parent

alabama_banks = pd.read_csv(BASE_DIR / "datos" / "alabama_banks.csv")
arkansas_banks = pd.read_csv(BASE_DIR / "datos" / "arkansas_banks.csv")
california_banks = pd.read_csv(BASE_DIR / "datos" / "california_banks.csv")
florida_banks = pd.read_csv(BASE_DIR / "datos" / "florida_banks.csv")
georgia_banks = pd.read_csv(BASE_DIR / "datos" / "georgia_banks.csv")
mississippi_banks = pd.read_csv(BASE_DIR / "datos" / "mississippi_banks.csv")
new_york_banks = pd.read_csv(BASE_DIR / "datos" / "new_york_banks.csv")
north_carolina_banks = pd.read_csv(BASE_DIR / "datos" / "north_carolina_banks.csv")
south_carolina_banks = pd.read_csv(BASE_DIR / "datos" / "south_carolina_banks.csv")
texas_banks = pd.read_csv(BASE_DIR / "datos" / "texas_banks.csv")
```
